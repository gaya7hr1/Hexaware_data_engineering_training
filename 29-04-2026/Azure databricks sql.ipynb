```python
spark.sql("""
CREATE OR REPLACE TABLE employees (
emp_id INT,
emp_name STRING,
city STRING,
department STRING,
salary INT,
experience INT
)
""")
```

```python
spark.sql("""
CREATE OR REPLACE TABLE projects (
project_id INT,
project_name STRING,
department STRING,
budget INT
)
""")
```

```python
spark.sql("SELECT * FROM employees").show()

spark.sql("SELECT emp_name, city, salary FROM employees").show()

spark.sql("SELECT DISTINCT department FROM employees").show()

spark.sql("SELECT COUNT(*) FROM employees").show()

spark.sql("SELECT * FROM employees WHERE city='Hyderabad'").show()

spark.sql("SELECT * FROM employees WHERE salary > 90000").show()
```

```python
spark.sql("SELECT * FROM employees ORDER BY salary ASC").show()

spark.sql("SELECT * FROM employees ORDER BY salary DESC LIMIT 5").show()
```

```python
spark.sql("SELECT AVG(salary) FROM employees").show()

spark.sql("SELECT department, COUNT(*) FROM employees GROUP BY department").show()
```

```python
spark.sql("SELECT department, SUM(salary) FROM employees GROUP BY department").show()

spark.sql("SELECT city, COUNT(*) FROM employees GROUP BY city").show()
```

```python
spark.sql("""
SELECT * FROM employees
WHERE salary > (SELECT AVG(salary) FROM employees)
""").show()
```

```python
spark.sql("""
SELECT * FROM employees
WHERE department='Engineering'
ORDER BY salary DESC
LIMIT 3
""").show()
```

```python
spark.sql("""
CREATE OR REPLACE TABLE employee_salary_analysis AS
SELECT 
    emp_name,
    city,
    department,
    salary,
    experience,
    CASE 
        WHEN salary >= 95000 THEN 'High'
        WHEN salary >= 80000 THEN 'Medium'
        ELSE 'Low'
    END AS salary_category,
    AVG(salary) OVER (PARTITION BY department) AS department_avg_salary
FROM employees
""")
```
