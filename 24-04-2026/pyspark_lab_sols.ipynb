```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.window import Window

spark = SparkSession.builder.appName("Ecommerce").getOrCreate()
sc = spark.sparkContext
```

```python
customers = spark.read.csv("customers.csv", header=True, inferSchema=True)
products = spark.read.csv("products.csv", header=True, inferSchema=True)
orders = spark.read.csv("orders.csv", header=True, inferSchema=True)
order_items = spark.read.csv("order_items.csv", header=True, inferSchema=True)
payments = spark.read.csv("payments.csv", header=True, inferSchema=True)
```

## Section 1 — DataFrame Basics

```python
# 1. Show all customers
customers.show()

# 2. Print schema of customers
customers.printSchema()

# 3. Count total customers
customers.count()

# 4. Show first 5 customers
customers.show(5)

# 5. Display only name and city
customers.select("name", "city").show()

# 6. Count number of products
products.count()

# 7. Show product names and price
products.select("product_name", "price").show()

# 8. Display order table
orders.show()

# 9. Count total orders
orders.count()

# 10. Show all payment records
payments.show()
```

## Section 2 — Filtering Data

```python
# 11. Show customers from Hyderabad
customers.filter(col("city") == "Hyderabad").show()

# 12. Show customers older than 30
customers.filter(col("age") > 30).show()

# 13. Show products with price > 10000
products.filter(col("price") > 10000).show()

# 14. Show electronics products
products.filter(col("category") == "Electronics").show()

# 15. Show orders with status Delivered
orders.filter(col("status") == "Delivered").show()

# 16. Show cancelled orders
orders.filter(col("status") == "Cancelled").show()

# 17. Show payments made using UPI
payments.filter(col("payment_type") == "UPI").show()

# 18. Show customers from Bangalore and Delhi
customers.filter(col("city").isin("Bangalore", "Delhi")).show()

# 19. Show products costing less than 1000
products.filter(col("price") < 1000).show()

# 20. Show customers aged between 25 and 35
customers.filter((col("age") >= 25) & (col("age") <= 35)).show()
```

## Section 3 — Selecting and Renaming

```python
# 21. Rename column price to product_price
products = products.withColumnRenamed("price", "product_price")
products.show()

# 22. Rename name to customer_name
customers = customers.withColumnRenamed("name", "customer_name")
customers.show()

# 23. Select customer name and age
customers.select("customer_name", "age").show()

# 24. Select product name and category
products.select("product_name", "category").show()

# 25. Select order_id and order_date
orders.select("order_id", "order_date").show()

# 26. Select payment type and amount
payments.select("payment_type", "amount").show()

# 27. Create column price_in_lakhs
products = products.withColumn("price_in_lakhs", col("product_price") / 100000)
products.select("product_name", "product_price", "price_in_lakhs").show()

# 28. Create column age_group
customers = customers.withColumn(
    "age_group",
    when(col("age") < 30, "Young")
    .when((col("age") >= 30) & (col("age") <= 40), "Middle")
    .otherwise("Senior")
)
customers.select("customer_name", "age", "age_group").show()

# 29. Select order items with quantity > 2
order_items.filter(col("quantity") > 2).show()

# 30. Display customers sorted by age
customers.orderBy("age").show()
```

## Section 4 — Sorting and Limiting

```python
# 31. Sort customers by age ascending
customers.orderBy(col("age").asc()).show()

# 32. Sort customers by age descending
customers.orderBy(col("age").desc()).show()

# 33. Show top 5 expensive products
products.orderBy(col("product_price").desc()).show(5)

# 34. Show cheapest 3 products
products.orderBy(col("product_price").asc()).show(3)

# 35. Sort orders by order_date
orders.orderBy("order_date").show()

# 36. Sort payments by amount
payments.orderBy("amount").show()

# 37. Show top 3 largest payments
payments.orderBy(col("amount").desc()).show(3)

# 38. Show smallest payments
payments.orderBy(col("amount").asc()).show(3)

# 39. Sort customers by city
customers.orderBy("city").show()

# 40. Sort products by category
products.orderBy("category").show()
```

## Section 5 — Aggregations

```python
# 41. Find total revenue from payments
payments.agg(sum("amount").alias("total_revenue")).show()

# 42. Find average payment amount
payments.agg(avg("amount").alias("avg_payment")).show()

# 43. Find max payment amount
payments.agg(max("amount").alias("max_payment")).show()

# 44. Find min payment amount
payments.agg(min("amount").alias("min_payment")).show()

# 45. Count customers by city
customers.groupBy("city").count().orderBy("city").show()

# 46. Count products by category
products.groupBy("category").count().show()

# 47. Average price per category
products.groupBy("category").agg(avg("product_price").alias("avg_price")).show()

# 48. Total quantity sold per product
order_items.groupBy("product_id").agg(sum("quantity").alias("total_qty")).show()

# 49. Average customer age
customers.agg(avg("age").alias("avg_age")).show()

# 50. Find total orders
orders.count()
```

## Section 6 — Joins

```python
# 51. Join orders with customers
orders_customers = orders.join(customers, "customer_id")
orders_customers.show()

# 52. Show customer name and order status
orders_customers.select("customer_name", "status").show()

# 53. Join orders with order_items
orders_items_joined = orders.join(order_items, "order_id")
orders_items_joined.show()

# 54. Join order_items with products
items_products = order_items.join(products, "product_id")
items_products.show()

# 55. Create column revenue = quantity * price
items_products = items_products.withColumn("revenue", col("quantity") * col("product_price"))
items_products.select("order_id", "product_name", "quantity", "product_price", "revenue").show()

# 56. Join orders, customers, products
full_join = orders.join(order_items, "order_id")                   .join(products, "product_id")                   .join(customers, "customer_id")
full_join.show()

# 57. Show customer name, product name, quantity
full_join.select("customer_name", "product_name", "quantity").show()

# 58. Calculate revenue per order
revenue_per_order = items_products.groupBy("order_id")     .agg(sum("revenue").alias("order_revenue"))
revenue_per_order.show()

# 59. Calculate revenue per product
revenue_per_product = items_products.groupBy("product_id", "product_name")     .agg(sum("revenue").alias("product_revenue"))
revenue_per_product.orderBy(col("product_revenue").desc()).show()

# 60. Calculate revenue per customer
revenue_per_customer = full_join.withColumn("revenue", col("quantity") * col("product_price"))     .groupBy("customer_id", "customer_name")     .agg(sum("revenue").alias("customer_revenue"))
revenue_per_customer.orderBy(col("customer_revenue").desc()).show()
```

## Section 7 — GroupBy Analytics

```python
# 61. Count orders per customer
orders.groupBy("customer_id").count().alias("order_count").show()

# 62. Count orders per city
orders.join(customers, "customer_id")     .groupBy("city").count().orderBy("city").show()

# 63. Revenue per product
items_products.groupBy("product_id", "product_name")     .agg(sum("revenue").alias("revenue")).orderBy(col("revenue").desc()).show()

# 64. Revenue per category
items_products.groupBy("category")     .agg(sum("revenue").alias("revenue")).show()

# 65. Revenue per city
full_join.withColumn("revenue", col("quantity") * col("product_price"))     .groupBy("city").agg(sum("revenue").alias("revenue")).orderBy(col("revenue").desc()).show()

# 66. Quantity sold per product
order_items.groupBy("product_id").agg(sum("quantity").alias("total_qty")).show()

# 67. Quantity sold per category
items_products.groupBy("category").agg(sum("quantity").alias("total_qty")).show()

# 68. Revenue per customer
revenue_per_customer.show()

# 69. Top 5 customers by revenue
revenue_per_customer.orderBy(col("customer_revenue").desc()).show(5)

# 70. Top 3 products by revenue
items_products.groupBy("product_id", "product_name")     .agg(sum("revenue").alias("revenue"))     .orderBy(col("revenue").desc()).show(3)
```

## Section 8 — Window Functions

```python
# 71. Rank products by price
w_global = Window.orderBy(col("product_price").desc())
products.withColumn("price_rank", rank().over(w_global)).show()

# 72. Rank products within category
w_cat = Window.partitionBy("category").orderBy(col("product_price").desc())
products.withColumn("rank_in_category", rank().over(w_cat)).show()

# 73. Row number for customers by signup_date
w_signup = Window.orderBy("signup_date")
customers.withColumn("row_num", row_number().over(w_signup)).show()

# 74. Rank customers by revenue
w_rev = Window.orderBy(col("customer_revenue").desc())
revenue_per_customer.withColumn("revenue_rank", rank().over(w_rev)).show()

# 75. Dense rank customers by revenue
revenue_per_customer.withColumn("dense_revenue_rank", dense_rank().over(w_rev)).show()

# 76. Find top customer per city
w_city = Window.partitionBy("city").orderBy(col("customer_revenue").desc())
top_per_city = revenue_per_customer.join(customers.select("customer_id","city"), "customer_id")     .withColumn("city_rank", rank().over(w_city))     .filter(col("city_rank") == 1)
top_per_city.show()

# 77. Find top product per category
w_cat_rev = Window.partitionBy("category").orderBy(col("revenue").desc())
items_products.groupBy("product_id", "product_name", "category")     .agg(sum("revenue").alias("revenue"))     .withColumn("cat_rank", rank().over(w_cat_rev))     .filter(col("cat_rank") == 1).show()

# 78. Running total revenue by order date
w_date = Window.orderBy("order_date").rowsBetween(Window.unboundedPreceding, Window.currentRow)
revenue_per_order.join(orders.select("order_id","order_date"), "order_id")     .withColumn("running_total", sum("order_revenue").over(w_date))     .orderBy("order_date").show()

# 79. Running total revenue by product
w_prod = Window.orderBy("product_id").rowsBetween(Window.unboundedPreceding, Window.currentRow)
items_products.groupBy("product_id", "product_name")     .agg(sum("revenue").alias("revenue"))     .withColumn("running_total", sum("revenue").over(w_prod))     .orderBy("product_id").show()

# 80. Rank orders by payment amount
w_pay = Window.orderBy(col("amount").desc())
payments.withColumn("payment_rank", rank().over(w_pay)).show()
```

## Section 9 — RDD Exercises

```python
logs = sc.textFile("logs.txt")

# 81. Count log records
print("Total log records:", logs.count())

# 82. Extract usernames
usernames = logs.map(lambda line: line.split()[1])
print("Usernames:", usernames.collect())

# 83. Extract actions
actions = logs.map(lambda line: line.split()[0])
print("Actions:", actions.collect())

# 84. Find unique users
unique_users = usernames.distinct()
print("Unique users:", unique_users.collect())

# 85. Count login events
login_count = logs.filter(lambda line: line.startswith("login")).count()
print("Login events:", login_count)

# 86. Count purchase events
purchase_count = logs.filter(lambda line: line.startswith("purchase")).count()
print("Purchase events:", purchase_count)

# 87. Count view events
view_count = logs.filter(lambda line: line.startswith("view")).count()
print("View events:", view_count)

# 88. Create (user, 1) pairs
user_pairs = logs.map(lambda line: (line.split()[1], 1))
print("User pairs sample:", user_pairs.take(5))

# 89. Use reduceByKey to count activity per user
activity_count = user_pairs.reduceByKey(lambda a, b: a + b)
print("Activity per user:", activity_count.collect())

# 90. Find most active user
most_active = activity_count.sortBy(lambda x: x[1], ascending=False).first()
print("Most active user:", most_active)
```

## Section 10 — Spark SQL

```python
# Create temp views
customers.createOrReplaceTempView("customers")
orders.createOrReplaceTempView("orders")
products.createOrReplaceTempView("products")
payments.createOrReplaceTempView("payments")
order_items.createOrReplaceTempView("order_items")

# 91. Show all customers using SQL
spark.sql("SELECT * FROM customers").show()

# 92. Find customers older than 30
spark.sql("SELECT * FROM customers WHERE age > 30").show()

# 93. Count customers per city
spark.sql("SELECT city, COUNT(*) AS customer_count FROM customers GROUP BY city ORDER BY city").show()

# 94. Average age per city
spark.sql("SELECT city, AVG(age) AS avg_age FROM customers GROUP BY city ORDER BY city").show()

# 95. Join customers and orders
spark.sql("SELECT * FROM customers c JOIN orders o ON c.customer_id = o.customer_id").show()

# 96. Show orders per customer
spark.sql("""
    SELECT c.customer_name, COUNT(o.order_id) AS order_count
    FROM customers c
    JOIN orders o ON c.customer_id = o.customer_id
    GROUP BY c.customer_name
    ORDER BY order_count DESC
""").show()

# 97. Join order_items and products
spark.sql("SELECT * FROM order_items oi JOIN products p ON oi.product_id = p.product_id").show()

# 98. Calculate revenue using SQL
spark.sql("""
    SELECT oi.order_id, p.product_name, oi.quantity, p.product_price,
           (oi.quantity * p.product_price) AS revenue
    FROM order_items oi
    JOIN products p ON oi.product_id = p.product_id
""").show()

# 99. Find top 3 customers by revenue
spark.sql("""
    SELECT c.customer_name, SUM(oi.quantity * p.product_price) AS total_revenue
    FROM orders o
    JOIN order_items oi ON o.order_id = oi.order_id
    JOIN products p ON oi.product_id = p.product_id
    JOIN customers c ON o.customer_id = c.customer_id
    GROUP BY c.customer_name
    ORDER BY total_revenue DESC
    LIMIT 3
""").show()

# 100. Find top product by revenue
spark.sql("""
    SELECT p.product_name, SUM(oi.quantity * p.product_price) AS total_revenue
    FROM order_items oi
    JOIN products p ON oi.product_id = p.product_id
    GROUP BY p.product_name
    ORDER BY total_revenue DESC
    LIMIT 1
""").show()
```

## Final Capstone Mini Project

```python
# 1. Join all tables
final_df = orders.join(order_items, "order_id").join(products, "product_id").join(customers, "customer_id").join(payments.select("order_id", "payment_type"), "order_id")

# 2. Calculate revenue
final_df = final_df.withColumn("revenue", col("quantity") * col("product_price"))
final_df.show()

# 3. Find top customers
top_customers = final_df.groupBy("customer_id", "customer_name").agg(sum("revenue").alias("total_revenue")).orderBy(col("total_revenue").desc())
top_customers.show()

# 4. Find top products
top_products = final_df.groupBy("product_id", "product_name").agg(sum("revenue").alias("total_revenue")).orderBy(col("total_revenue").desc())
top_products.show()

# 5. Find revenue per city
revenue_city = final_df.groupBy("city").agg(sum("revenue").alias("city_revenue")).orderBy(col("city_revenue").desc())
revenue_city.show()

# 6. Rank customers by revenue
w_rank = Window.orderBy(col("total_revenue").desc())
ranked_customers = top_customers.withColumn("revenue_rank", rank().over(w_rank))
ranked_customers.show()

# 7. Save final dataset
final_df.write.mode("overwrite").csv("ecommerce_report", header=True)
print("Final dataset saved to ecommerce_report")
```
