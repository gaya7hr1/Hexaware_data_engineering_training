```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("example").getOrCreate()
```

Exercise 2 — Load employees.csv

```python
emp_df=spark.read.csv("employees.csv",header=True,inferSchema=True)
```

Exercise 3 — Load departments.csv

```python
dept_df = spark.read.csv("departments.csv", header=True, inferSchema=True)
```

Exercise 4 — Load sales.csv

```python
sales_df = spark.read.csv("sales.csv", header=True, inferSchema=True)
```

Exercise 5 — Load employees_nested.json

```python
nested_df = spark.read.json("employees_nested.json")
```

Exercise 6 — Print Schema

```python
emp_df.printSchema()
dept_df.printSchema()
sales_df.printSchema()
nested_df.printSchema()
```

Exercise 7 — Show first 5 rows

```python
emp_df.show(5)
```

Exercise 8 — Count total employees

```python
emp_df.count()
```

Exercise 9 — List all column names

```python
emp_df.columns
```

Exercise 10 — Show only 3 rows

```python
emp_df.show(3)
```

Exercise 11 — Select name and salary

```python
emp_df.select("name", "salary").show()
```

Exercise 12 — Rename salary to emp_salary

```python
emp_df.withColumnRenamed("salary", "emp_salary").show()
```

Exercise 13 — Employees joined after 2023-01-01

```python
from pyspark.sql.functions import col

emp_df.filter(col("joining_date") > "2023-01-01").show()
```

Exercise 14 — Salary > 65000

```python
emp_df.filter(col("salary") > 65000).show()
```

Exercise 15 — Employees in IT

```python
emp_df.filter(col("department") == "IT").show()
```

Exercise 16 — IT & salary > 70000

```python
emp_df.filter((col("department") == "IT") & (col("salary") > 70000)).show()
```

Exercise 17 — Drop joining_date

```python
emp_df.drop("joining_date").show()
```

Exercise 18 — Sort salary ascending

```python
emp_df.orderBy("salary").show()
```

Exercise 19 — Sort salary descending

```python
emp_df.orderBy(col("salary").desc()).show()
```

Exercise 20 — Top 3 highest paid

```python
emp_df.orderBy(col("salary").desc()).limit(3).show()
```

Exercise 21 — Total salary

```python
from pyspark.sql.functions import sum

emp_df.select(sum("salary")).show()
```

Exercise 22 — Average salary

```python
from pyspark.sql.functions import avg

emp_df.select(avg("salary")).show()
```

Exercise 23 — Highest salary

```python
from pyspark.sql.functions import max

emp_df.select(max("salary")).show()
```

Exercise 24 — Lowest salary

```python
from pyspark.sql.functions import min

emp_df.select(min("salary")).show()
```

Exercise 25 — Count employees per department

```python
emp_df.groupBy("department").count().show()
```

Exercise 26 — Average salary per department

```python

emp_df.groupBy("department").agg(avg("salary")).show()
```

Exercise 27 — Total salary per department

```python

emp_df.groupBy("department").agg(sum("salary")).show()
```

Exercise 28 — Inner join

```python
emp_df.join(dept_df, on="department", how="inner").show()
```

Exercise 29 — Left join

```python
emp_df.join(dept_df, on="department", how="left").show()
```

Exercise 30 — Employee name with location

```python
emp_df.join(dept_df, on="department", how="left") \
      .select("name", "location") \
      .show()
```

Exercise 31 — Add bonus column (10%)

```python


emp_df.withColumn("bonus", col("salary") * 0.10).show()
```

Exercise 32 — Add company column

```python
from pyspark.sql.functions import lit

emp_df.withColumn("company", lit("BotCampus")).show()
```

Exercise 33 — Rows with null department

```python
emp_df.filter(col("department").isNull()).show()
```

Exercise 34 — Replace null with "Unknown"

```python
emp_df.fillna({"department": "Unknown"}).show()
```

Exercise 35 — Drop rows with nulls

```python
emp_df.dropna().show()
```

Exercise 36 — Names to uppercase

```python
from pyspark.sql.functions import upper

emp_df.select(upper("name")).show()
```

Exercise 37 — Length of names

```python
from pyspark.sql.functions import length

emp_df.select("name", length("name")).show()
```

Exercise 38 — Salary in lakhs

```python
emp_df.withColumn("salary_in_lakhs", col("salary") / 100000).show()
```
