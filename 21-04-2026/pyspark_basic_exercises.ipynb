### Exercise 1

Create a DataFrame with:

```
Name
Salary
Department
```

Add 5 employees.

---

````python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('PySpark Basics').getOrCreate()
print("spark session created")
````

````python
data = [
    ("Alice", 50000, "HR"),
    ("Bob", 60000, "IT"),
    ("Charlie", 55000, "Finance"),
    ("David", 70000, "IT"),
    ("Eve", 65000, "Marketing")
]

columns = ["Name", "Salary", "Department"]

df = spark.createDataFrame(data, columns)
df.show()
````

### Exercise 2

Show only **Name and Salary** columns.

---

````python
df.select("Name", "Salary").show()
````

### Exercise 3

Filter employees with **salary greater than 50000**.

---

````python
df.filter(df.Salary > 50000).show()
````

### Exercise 4

Count employees per department.

---

````python
df.groupBy("Department").count().show()
````

## Exercise 5

Sort employees by salary.

---


````python
df.orderBy("Salary").show()
````
