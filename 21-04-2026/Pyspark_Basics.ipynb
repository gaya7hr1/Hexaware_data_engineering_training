```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('PySpark Basics').getOrCreate()
print("spark session created")
```

```python
print(spark.version)
```

```python
#pd we use dict list , spark uses list of tuples
data=[
    ("Rahul",25,"Hyd"),
    ("Sneha",28,"Banglore"),
    ("Arjun",24,"Chennai"),
    ("Priya",30,"Mumbai")
]
columns=["Name","Age","City"]
df=spark.createDataFrame(data,columns)
df.show()
```

```python
df.printSchema()
```

```python
df.select("Name").show()
```

```python
df.select("Name","City").show()
```

```python
df.filter(df.Age>25).show()
```

```python
df.groupBy("City").count().show()
```

```python
df.orderBy("Age").show()
```

```python
df.orderBy(df.Age.desc()).show()
```
