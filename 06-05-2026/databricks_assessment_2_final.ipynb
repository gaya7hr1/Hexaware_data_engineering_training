```python
from pyspark.sql.functions import *

orders_df = spark.createDataFrame(orders_data, orders_columns)
orders_df.show()
#11
orders_df = orders_df.withColumn("order_date",to_date(col("order_date"),"yyyy-MM-dd"))
orders_df.show()
#12
orders_df = orders_df.join(products_df.select("product_id","price"),"product_id").withColumn("total_order_value",col("quantity")*col("price"))
orders_df.show()
#13
products_df = products_df.withColumn("stock_status",when(col("stock_quantity")<20,"Low").otherwise("Sufficient"))
orders_df.show()
#14
orders_df = orders_df.withColumn("order_priority",when(col("quantity")>=50,"High").when(col("quantity")>=10,"Medium").otherwise("Low"))
orders_df.show()
#15
products_df = products_df.withColumn("expensive_product_flag",when(col("price")>20000,"Yes").otherwise("No"))
products_df.show()
#16
products_df = products_df.withColumn("category",upper(col("category")))
products_df.show()
#17
products_df = products_df.withColumnRenamed("warehouse_city","inventory_city")
products_df.show()
#18
products_df = products_df.withColumn("inventory_value",col("price")*col("stock_quantity"))
products_df.show()
#19
orders_df = orders_df.drop("price")
orders_df.show()
#20
products_df = products_df.withColumn("low_stock_flag",when(col("stock_quantity")<20,"Yes").otherwise("No"))
products_df.show()
```

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import *

#51
window1 = Window.partitionBy("category").orderBy(col("bill_amount").desc())
final_joined_df.withColumn("rank",rank().over(window1)).select("category","product_name","bill_amount","rank").show()

#52
window2 = Window.partitionBy("supplier_city").orderBy(col("bill_amount").desc())
final_joined_df.withColumn("rank",rank().over(window2)).select("supplier_city","supplier_name","bill_amount","rank").show()

#53
window3 = Window.partitionBy("category").orderBy(col("bill_amount").desc())
final_joined_df.withColumn("row_num",row_number().over(window3)).filter(col("row_num")==1).show()

#54
window4 = Window.orderBy(col("bill_amount").desc())
final_joined_df.withColumn("dense_rank",dense_rank().over(window4)).select("supplier_name","bill_amount","dense_rank").show()

#55
supplier_revenue_df = final_joined_df.groupBy("supplier_name").agg(sum("bill_amount").alias("total_revenue"))
window5 = Window.orderBy(col("total_revenue").desc())
supplier_revenue_df.withColumn("rank",dense_rank().over(window5)).filter(col("rank")<=2).show()

#56
window6 = Window.orderBy("order_date")
final_joined_df.withColumn("running_total",sum("bill_amount").over(window6)).select("order_date","bill_amount","running_total").show()

#57
window7 = Window.partitionBy("supplier_name").orderBy("order_date")
final_joined_df.withColumn("running_total",sum("bill_amount").over(window7)).select("supplier_name","order_date","bill_amount","running_total").show()

#58
city_df = final_joined_df.groupBy("supplier_city").agg(sum("bill_amount").alias("total_revenue"))
window8 = Window.orderBy(col("total_revenue").desc())
city_df.withColumn("rank",rank().over(window8)).show()

#59
category_df = final_joined_df.groupBy("category").agg(sum("bill_amount").alias("total_revenue"))
window9 = Window.orderBy(col("total_revenue").desc())
category_df.withColumn("rank",rank().over(window9)).show()

#60
window10 = Window.partitionBy("payment_mode").orderBy(col("bill_amount").desc())
final_joined_df.withColumn("row_num",row_number().over(window10)).filter(col("row_num")==1).show()
```

```python
from delta.tables import DeltaTable
from pyspark.sql.functions import *

#71
orders_df.write.format("delta").mode("overwrite").save("/tmp/delta/orders_target")

#72
target_df = spark.read.format("delta").load("/tmp/delta/orders_target")
target_df.show()

#73
daily_orders_df.createOrReplaceTempView("daily_orders_updates")

#74
delta_table = DeltaTable.forPath(spark,"/tmp/delta/orders_target")

delta_table.alias("t").merge(
    daily_orders_df.alias("s"),
    "t.order_id = s.order_id"
).whenMatchedUpdate(set = {
    "product_id":"s.product_id",
    "supplier_id":"s.supplier_id",
    "order_date":"s.order_date",
    "quantity":"s.quantity",
    "order_status":"s.order_status"
}).whenNotMatchedInsert(values = {
    "order_id":"s.order_id",
    "product_id":"s.product_id",
    "supplier_id":"s.supplier_id",
    "order_date":"s.order_date",
    "quantity":"s.quantity",
    "order_status":"s.order_status"
}).execute()

#75
delta_table.toDF().filter(col("order_id").isin(304,319)).show()

#76
delta_table.toDF().filter(col("order_id").isin(321,322,323)).show()

#77
delta_table.toDF().filter(col("order_id").isin(304,319)).select("order_id","order_status").show()

#78
delta_table.toDF().filter(col("order_id").isin(321,322,323)).select("order_id").show()

#79
spark.sql("DESCRIBE HISTORY delta.`/tmp/delta/orders_target`").show(truncate=False)

#80 MERGE is important because it performs update and insert together efficiently using UPSERT logic
```
