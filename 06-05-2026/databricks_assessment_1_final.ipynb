# part 2

```python

from pyspark.sql.functions import *
from pyspark.sql.types import *

# 11. Convert registration_date into date type
patients_df = patients_df.withColumn(
    "registration_date",
    to_date("registration_date")
)

# 12. Convert visit_date into date type
visits_df = visits_df.withColumn(
    "visit_date",
    to_date("visit_date")
)

# 13. Add a column test_cost
visits_df = visits_df.withColumn(
    "test_cost",
    col("tests_count") * 500
)

# 14. Add a column estimated_total_bill
visits_doctors_df = visits_df.join(
    doctors_df,
    "doctor_id"
)

visits_doctors_df = visits_doctors_df.withColumn(
    "estimated_total_bill",
    col("consultation_fee") + col("test_cost")
)

# 15. Create age_category
patients_df = patients_df.withColumn(
    "age_category",
    when(col("age") < 30, "Young")
    .when(col("age") < 45, "Adult")
    .otherwise("Senior")
)

# 16. Create membership_priority
patients_df = patients_df.withColumn(
    "membership_priority",
    when(col("membership") == "Gold", 1)
    .when(col("membership") == "Silver", 2)
    .otherwise(3)
)

# 17. Create visit_priority
visits_df = visits_df.withColumn(
    "visit_priority",
    when(col("tests_count") >= 3, "High")
    .otherwise("Normal")
)

# 18. Create high_bill_flag
visits_doctors_df = visits_doctors_df.withColumn(
    "high_bill_flag",
    when(col("estimated_total_bill") > 3000, "Yes")
    .otherwise("No")
)

# 19. Rename patient_name to full_name
patients_df = patients_df.withColumnRenamed(
    "patient_name",
    "full_name"
)

# 20. Drop any temporary column
visits_doctors_df = visits_doctors_df.drop("test_cost")
```

# part 6

```python

from pyspark.sql.window import Window
from pyspark.sql.functions import *

# 51. Rank patients by total bill within city
patient_bill_df = final_df.groupBy(
    "city",
    "full_name"
).agg(
    sum("bill_amount").alias("total_bill")
)

window_city = Window.partitionBy("city").orderBy(col("total_bill").desc())

patient_bill_df.withColumn(
    "rank",
    rank().over(window_city)
).show()

# 52. Rank doctors by total revenue within specialization
doctor_revenue_df = final_df.groupBy(
    "specialization",
    "doctor_name"
).agg(
    sum("bill_amount").alias("total_revenue")
)

window_spec = Window.partitionBy("specialization").orderBy(col("total_revenue").desc())

doctor_revenue_df.withColumn(
    "rank",
    rank().over(window_spec)
).show()

# 53. ROW_NUMBER for top billing patient per city
patient_bill_df.withColumn(
    "row_num",
    row_number().over(window_city)
).filter(
    col("row_num") == 1
).show()

# 54. DENSE_RANK doctors by consultation fee within specialization
window_fee = Window.partitionBy("specialization").orderBy(col("consultation_fee").desc())

doctors_df.withColumn(
    "dense_rank",
    dense_rank().over(window_fee)
).show()

# 55. Top 2 doctors by revenue
doctor_total_df = final_df.groupBy(
    "doctor_name"
).agg(
    sum("bill_amount").alias("total_revenue")
)

window_doc = Window.orderBy(col("total_revenue").desc())

doctor_total_df.withColumn(
    "rank",
    dense_rank().over(window_doc)
).filter(
    col("rank") <= 2
).show()

# 56. Running total revenue by visit date
date_revenue_df = final_df.groupBy(
    "visit_date"
).agg(
    sum("bill_amount").alias("daily_revenue")
)

window_date = Window.orderBy("visit_date")

date_revenue_df.withColumn(
    "running_total",
    sum("daily_revenue").over(window_date)
).show()

# 57. Running total revenue by doctor
doctor_running_df = final_df.groupBy(
    "doctor_name"
).agg(
    sum("bill_amount").alias("doctor_revenue")
)

window_running = Window.orderBy("doctor_revenue")

doctor_running_df.withColumn(
    "running_total",
    sum("doctor_revenue").over(window_running)
).show()

# 58. Rank cities by total revenue
city_revenue_df = final_df.groupBy(
    "city"
).agg(
    sum("bill_amount").alias("total_revenue")
)

window_city_rank = Window.orderBy(col("total_revenue").desc())

city_revenue_df.withColumn(
    "rank",
    rank().over(window_city_rank)
).show()

# 59. Rank specializations by total revenue
spec_revenue_df = final_df.groupBy(
    "specialization"
).agg(
    sum("bill_amount").alias("total_revenue")
)

window_spec_rank = Window.orderBy(col("total_revenue").desc())

spec_revenue_df.withColumn(
    "rank",
    rank().over(window_spec_rank)
).show()

# 60. Highest bill visit per payment mode
window_payment = Window.partitionBy("payment_mode").orderBy(col("bill_amount").desc())

final_df.withColumn(
    "row_num",
    row_number().over(window_payment)
).filter(
    col("row_num") == 1
).show()
```

```python

from delta.tables import DeltaTable
from pyspark.sql.functions import *

# 61. Save final joined data as Delta table
final_df.write.format("delta").mode("overwrite").saveAsTable("hospital_final_delta")

# 62. Insert two new visit records
new_visits = [
(26,1002,201,"2024-03-14","Completed",2),
(27,1005,210,"2024-03-14","Completed",1)
]

new_visits_df = spark.createDataFrame(
    new_visits,
    visits_columns
)

new_visits_df.write.format("delta") .mode("append").saveAsTable("hospital_visits_delta")

# 63. Update visit status for one pending visit
delta_visits = DeltaTable.forName(spark, "hospital_visits_delta")

delta_visits.update(
    condition = "visit_id = 11",
    set = {
        "visit_status": "'Completed'"
    }
)
payments_df.write.format("delta") \
.mode("overwrite") \
.saveAsTable("hospital_payments_delta")
# 64. Update payment status for one pending payment
delta_payments = DeltaTable.forName(spark, "hospital_payments_delta")

delta_payments.update(
    condition = "payment_id = 311",
    set = {
        "payment_status": "'Paid'"
    }
)

# 65. Delete cancelled visits
delta_visits.delete(
    condition = "visit_status = 'Cancelled'"
)

# 66. Create cleaned completed visits Delta table
completed_df = visits_df.filter(
    col("visit_status") == "Completed"
)

completed_df.write.format("delta") .mode("overwrite") .saveAsTable("completed_visits_delta")

# 67. Query Delta table history
spark.sql("""
DESCRIBE HISTORY hospital_visits_delta
""").show(truncate=False)

# 68. Query older version using time travel
spark.read.format("delta") .option("versionAsOf", 0) .table("hospital_visits_delta") .show()

# 69. Compare latest and older version
latest_df = spark.read.format("delta").table("hospital_visits_delta")

old_df = spark.read.format("delta") .option("versionAsOf", 0).table("hospital_visits_delta")

latest_df.show()
old_df.show()

# 70. VACUUM DRY RUN
spark.sql("""
VACUUM hospital_visits_delta RETAIN 168 HOURS DRY RUN
""").show(truncate=False)
```

# part 8

```python

# 71. Create Delta target table
visits_df.write.format("delta").mode("overwrite") .saveAsTable("hospital_visits_target")

# 72. Initial visits loaded
spark.table("hospital_visits_target").show()

# 73. Create temp view for daily updates
daily_visits_df.createOrReplaceTempView("daily_visits_updates")

# 74,75,76. Merge updates and inserts
target_table = DeltaTable.forName(
    spark,
    "hospital_visits_target"
)

target_table.alias("target").merge(
    daily_visits_df.alias("source"),
    "target.visit_id = source.visit_id"
).whenMatchedUpdate(set={
    "patient_id": "source.patient_id",
    "doctor_id": "source.doctor_id",
    "visit_date": "source.visit_date",
    "visit_status": "source.visit_status",
    "tests_count": "source.tests_count"
}).whenNotMatchedInsert(values={
    "visit_id": "source.visit_id",
    "patient_id": "source.patient_id",
    "doctor_id": "source.doctor_id",
    "visit_date": "source.visit_date",
    "visit_status": "source.visit_status",
    "tests_count": "source.tests_count"
}).execute()

# 77. Verify updated visit IDs
spark.sql("""
SELECT *
FROM hospital_visits_target
WHERE visit_id IN (11,20)
""").show()

# 78. Verify newly inserted visit IDs
spark.sql("""
SELECT *
FROM hospital_visits_target
WHERE visit_id IN (26,27,28)
""").show()

# 79. Check Delta history
spark.sql("""
DESCRIBE HISTORY hospital_visits_target
""").show(truncate=False)

# 80. Why MERGE is used
#MERGE is used for incremental loads because it handles both updates and inserts in one operation.
```
