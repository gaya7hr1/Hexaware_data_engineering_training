```python
from pyspark.sql import SparkSession
spark=SparkSession.builder.appName('rdd').getOrCreate()
sc=spark.sparkContext
print("Spark ready")
```

```python
#parallel execution
numbers=[10,20,30,40,50]
rdd=sc.parallelize(numbers)
print(rdd)
```

```python
rdd.collect()
```

```python
data=["Rahul","Sneha","Arjun","Priya"]
name_rdd=sc.parallelize(data)
name_rdd.collect()
```

```python
#control the partitions
rdd.getNumPartitions()
```

```python
rdd=sc.parallelize(numbers,3)
rdd.getNumPartitions()
```

```python
rdd=sc.parallelize([10,20,30,40,50,60],3)
rdd.glom().collect()
```

RDD operations: two categories
1. Transformations
These create a new RDD.
Examples:
map,
filter,
flatMap,
distinct,
union,
intersection,
groupByKey,
reduceByKey
2. Actions
These return a result to the driver.
Examples:
collect,
count,
first,
take,
reduce

```python
rdd=sc.parallelize([1,2,3,4,5]) #lazy evaluation
mapped_rdd=rdd.map(lambda x:x*2)
filtered_rdd=mapped_rdd.filter(lambda x:x>5)
print("Nothing executed yet")
```

```python
filtered_rdd.collect()
```

map()

```python
rdd=sc.parallelize([1,2,3,4,5])
squared_rdd=rdd.map(lambda x:x*x)
squared_rdd.collect()
```

```python
rdd=sc.parallelize(["rahul","sneha","arjun"])
upper_rdd=rdd.map(lambda x:x.upper())
upper_rdd.collect()
```

filter()

```python
rdd=sc.parallelize(["apple","banana","mango","kiwi"])
long_words=rdd.filter(lambda x:len(x)>5)
long_words.collect()
```

flatMap()-catches multiple elements;
map()-catches one element

```python
lines=sc.parallelize([
    "python spark hadoop",
    "spark sql pyspark"
])
words=lines.flatMap(lambda line:line.split(" "))
words.collect()
```

```python
#same code using map() gives u the output in a nested list
lines=sc.parallelize([
    "python spark hadoop",
    "spark sql pyspark"
])
words=lines.map(lambda line:line.split(" "))
words.collect()
```

```python
rdd=sc.parallelize([10,20,20,30,30,30,40])
rdd.distinct().collect()
```

```python
rdd1=sc.parallelize([1,2,3])
rdd2=sc.parallelize([4,5,6])
rdd1.union(rdd2).collect()
```

```python
rdd1=sc.parallelize([1,2,3,4])
rdd2=sc.parallelize([3,4,5,6])
rdd1.intersection(rdd2).collect()
```

```python
rdd1=sc.parallelize([1,2,3,4])
rdd2=sc.parallelize([3,4])
rdd1.subtract(rdd2).collect()
```

```python
data=[
    ("IT",70000),
    ("HR",60000),
    ("IT",75000),
    ("Finance",80000),
    ("HR",50000)
]
rdd=sc.parallelize(data)
rdd.collect()
```

```python
rdd.reduceByKey(lambda a,b:a+b).collect()
```

```python
grouped=rdd.groupByKey()
[(k,list(v)) for k,v in grouped.collect()]
```

```python
rdd=sc.parallelize([
    ("Rahul",85),
    ("Sneha",92),
    ("Arjun",78)
])
rdd.mapValues(lambda x:x+5).collect()
```

```python
rdd=sc.parallelize([10,40,20,30])
rdd.sortBy(lambda x:x).collect()
```

```python
lines=sc.parallelize([
    "python spark python",
    "spark hadoop spark",
    "python hadoop"
])
word_counts=(
    lines.flatMap(lambda line:line.split(" "))
    .map(lambda word:(word,1))
    .reduceByKey(lambda a,b:a+b)
)
word_counts.collect()
```

```python
employees=sc.parallelize([
    ("Rahul",70000),
    ("Sneha",60000),
    ("Arjun",75000),
    ("Priya",80000),
    ("Karan",50000)
])
high_salary=employees.filter(lambda x:x[1]>65000)
high_salary.collect()
```
