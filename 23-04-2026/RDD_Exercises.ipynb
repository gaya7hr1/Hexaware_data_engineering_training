# Dataset 1 — Website Logs

```python
from pyspark.sql import SparkSession
spark=SparkSession.builder.appName('RDD Exercises').getOrCreate()
sc=spark.sparkContext
print("Spark ready")
```

```python
logs = [
"Rahul login",
"Sneha login",
"Arjun login",
"Rahul purchase",
"Priya login",
"Sneha logout",
"Rahul login",
"Karan login",
"Arjun purchase",
"Sneha login",
"Rahul logout",
"Amit login",
"Priya purchase",
"Karan purchase",
"Sneha login",
"Rahul purchase",
"Meera login",
"Arjun login",
"Sneha purchase",
"Rahul login"
]
logs_rdd = sc.parallelize(logs)
logs_rdd.collect()
```

 # Dataset 2 — Product Sales



```python
sales = [
("Laptop",1,75000),
("Mouse",3,500),
("Keyboard",2,1500),
("Laptop",1,75000),
("Monitor",1,12000),
("Mouse",2,500),
("Keyboard",1,1500),
("Laptop",1,75000),
("Monitor",2,12000),
("Mouse",3,500),
("Laptop",1,75000),
("Keyboard",2,1500),
("Monitor",1,12000),
("Mouse",1,500),
("Laptop",1,75000)
]
sales_rdd = sc.parallelize(sales)
sales_rdd.collect()
```

**Dataset 3 — Employee Data**

```python
employees = [
("Rahul","IT",70000),
("Sneha","HR",60000),
("Arjun","IT",75000),
("Priya","Finance",80000),
("Karan","IT",50000),
("Amit","HR",58000),
("Meera","Finance",82000),
("Ravi","IT",72000),
("Neha","HR",61000),

("Vikram","Finance",90000)
]
emp_rdd = sc.parallelize(employees)
emp_rdd.collect()
```

#  Part 2 — flatMap Practice

```python
sentences = sc.parallelize([
"big data spark",
"spark hadoop hive",
"python spark hadoop",
"spark streaming kafka"
])
```

```python

words=sentences.flatMap(lambda x:x.split(" "))
words.collect()
```

```python

words.distinct().collect()
```

```python

word_count=words.map(lambda word:(word,1)).reduceByKey(lambda a,b:a+b)
word_count.collect()
```

# Part 3 — Sales Dataset Exercises

```python

revenue=sales_rdd.map(lambda x:(x[0],x[1]*x[2]))
revenue.collect()
```

```python

total_revenue=revenue.map(lambda x: x[1]).reduce(lambda a,b:a+b)
print(total_revenue)
```

```python

sales_rdd.map(lambda x:x[0]).distinct().collect()
```

```python

sales_rdd.filter(lambda x:x[1]>1).collect()
```

```python

sales_qty=sales_rdd.map(lambda x:(x[0],x[1])).reduceByKey(lambda a,b:a+b)
sales_qty.collect()
```

```python

prod_rev=revenue.reduceByKey(lambda a,b:a+b)
prod_rev.collect()
```

```python

prod_rev.sortBy(lambda x:x[1],ascending=False).take(1)
```

# Part 4 — Employee Dataset

```python

emp_names=emp_rdd.map(lambda x:x[0])
emp_names.collect()
```

```python

emp_rdd.filter(lambda x:x[2]>70000).collect()
```

```python

emp_rdd.map(lambda x:x[1]).distinct().collect()
```

```python

dept_salary=emp_rdd.map(lambda x:(x[1],x[2])).reduceByKey(lambda a,b:a+b)
dept_salary.collect()
```

```python

avg_salary=emp_rdd.map(lambda x:(x[1],(x[2],1))).reduceByKey(lambda a,b: (a[0]+b[0], a[1]+b[1])).map(lambda x: (x[0],x[1][0]/x[1][1]))
avg_salary.collect()
```

```python

emp_rdd.sortBy(lambda x: x[2],ascending=False).take(1)
```
