# Section 7 — Data Cleaning

```python
#61
df=df.dropDuplicates()
df.show()
```

```python
#62
df.select([sum(col(c).isNull().cast("int")).alias(c) for c in df.columns]).show()
```

```python
#63
df=df.na.fill({"salary": 0})
df.show()
```

```python
#64
df=df.na.drop(subset=["department"])
df.show()
```

```python
#65
avg_exp=df.select(avg("experience")).collect()[0][0]
df=df.na.fill({"experience":avg_exp})
df.show()
```
